```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[6, 2],
              [2, 3]])
print("Square Matrix A: ")
print(A)
```

```python
eig_values, eig_vectors = np.linalg.eig(A)
print("Eigenvectors: ")
print(eig_vectors)
print("Eigenvalues: ",eig_values)
```

```python
v = eig_vectors[:, 1]
lambda_v = eig_values[1]
print("Eigenvector: ",v)
print("Eigenvalue: ",lambda_v)
```

```python
plt.figure(figsize=(6, 5))
origin = [0, 0]
plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1, color='blue', label="Original Eigenvector (v)")
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.axhline(0, color='grey', lw=1)
plt.axvline(0, color='grey', lw=1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()
plt.show()
```

```python
Av = np.dot(A,v)
print("Transformed vector: ",Av)
```

```python
plt.quiver(*origin, *Av, angles='xy', scale_units='xy', scale=1, color='red', label=f"Transformed (Av = λv, λ={lambda_v:.2f})")
plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1, color='blue', label="Original Eigenvector (v)")
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.axhline(0, color='grey', lw=1)
plt.axvline(0, color='grey', lw=1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()
plt.show()
```
